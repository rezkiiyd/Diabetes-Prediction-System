# Load The Dataset

```python
data = pd.read_csv('/content/diabetes.csv')
data
```

This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether a patient has diabetes based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

```python
data.dtypes
```

# Correlation Matrix

```python
correlation = data.corr()
print(correlation)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.show
```

Based on the heatmap, it appears that the variable â€˜Outcomeâ€™ has the strongest positive correlation with â€˜Glucoseâ€™. This suggests that higher glucose levels could be a strong indicator of diabetes outcome.

On the other hand, â€˜Blood Pressureâ€™ and â€˜Skin Thicknessâ€™ seem to have the weakest correlation with â€˜Outcomeâ€™. This implies that these variables might not be as significant in predicting the outcome of diabetes.

# Data Visualization

```python
sns.countplot(x='Pregnancies', hue='Outcome', data=data)
plt.title('Jumlah Kehamilan vs Diabetes')
plt.show()
```

ENG

The number of Negative Diabetes cases (blue bars) is higher for a lower number of pregnancies (especially at 1 pregnancy). The number of Positive Diabetes cases (orange bars) increases with the number of pregnancies but remains lower than Negative Diabetes until about six pregnancies, after which it becomes comparable or higher.

IDN

Jumlah Negative Diabetes (batang biru) lebih tinggi untuk jumlah kehamilan yang lebih rendah (terutama pada 1 kehamilan).
Jumlah Positive Diabetes (batang oranye) meningkat seiring dengan jumlah kehamilan tetapi tetap lebih rendah dari Negative Diabetes hingga sekitar enam kehamilan, setelah itu menjadi sebanding atau lebih tinggi.

```python
sns.scatterplot(x='BMI', y='SkinThickness', hue='Outcome', data=data)
plt.title('BMI vs SkinThickness (Dibagi berdasarkan Outcome)')
plt.show()
```

ENG

A large number of blue data points (Negative Diabetes) are gathered between the ages of approximately 20 to 50 and glucose levels are between about 75 and 125. The orange data points (Positive Diabetes) are more scattered but primarily appear in individuals aged around mid-30s to late 70s with glucose levels generally above about the midpoint of the y-axis. From this graph, it can be concluded that higher glucose levels associated with Positive Diabetes tend to be more common in older individuals, while lower glucose levels associated with Negative Diabetes are more common in younger individuals.

IDN

Sejumlah besar titik data biru (Negative Diabetes) terkumpul antara usia sekitar 20 hingga 50 dan kadar glukosa sekitar antara 75 dan 125.
Titik data oranye (Positive Diabetes) lebih tersebar tetapi terutama muncul pada individu yang berusia sekitar pertengahan 30-an hingga akhir 70-an dengan kadar glukosa umumnya di atas sekitar titik tengah sumbu y.
Dari grafik ini, dapat disimpulkan bahwa kadar glukosa yang lebih tinggi yang dikaitkan dengan Positive Diabetes cenderung lebih umum pada individu yang lebih tua, sementara kadar glukosa yang lebih rendah yang dikaitkan dengan Negative Diabetes lebih umum pada individu yang lebih muda.

```python
# Membuat histogram untuk BMI pasien diabetes
data[data['Outcome'] == 1]['Age'].hist(bins=30)
plt.title('Distribusi BMI untuk Pasien Diabetes')
plt.xlabel('Age')
plt.ylabel('Jumlah Pasien')
plt.show()
```

ENG

There are peaks visible around the age of 30 and just before the age of 50, indicating a higher number of patients with a certain BMI in these age groups. The output from this image shows that there is a variation in the number of diabetes patients with different BMIs at various ages. It seems there is a significant number of patients in their late twenties to early thirties and another peak just before the fifties. This could potentially indicate the ages at which people are more likely to have a certain BMI while being diabetic.

IDN

Ada puncak yang terlihat sekitar usia 30 dan tepat sebelum usia 50, menunjukkan jumlah pasien dengan BMI tertentu yang lebih tinggi di kelompok usia ini.
Output dari gambar ini menunjukkan bahwa ada variasi jumlah pasien diabetes dengan BMI yang berbeda di berbagai usia. Sepertinya ada jumlah pasien yang signifikan di akhir dua puluhan hingga awal tiga puluhan dan puncak lainnya tepat sebelum lima puluhan. Ini bisa berpotensi menunjukkan usia di mana orang lebih mungkin memiliki BMI tertentu sekaligus mengidap diabetes.

```python
import plotly.express as px
# Memfilter DataFrame untuk hanya pasien diabetes
df_diabetes = data[data['Outcome'] == 1]

# Membuat plot 3D
fig = px.scatter_3d(df_diabetes, x='Age', y='Glucose', z='BMI')
fig.update_layout(scene = dict(
                    xaxis_title='Usia',
                    yaxis_title='Glukosa',
                    zaxis_title='BMI'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=10))
fig.show()
```

ENG


**Data Distribution**: The majority of patient data is distributed in areas where the BMI ranges between 20 and 40, glucose levels are approximately between 100 and 150, and the age ranges from 20 to 60 years.

**Correlation**: There is no clear linear or direct correlation between BMI, age, and glucose levels among these diabetes patients. As age increases, there is no pattern indicating that BMI or glucose levels significantly increase or decrease.

**Patient Concentration**: Most patients have glucose levels between approximately 100 to 150. A large proportion of these patients are aged between approximately 20 to 60 years with a BMI ranging around 20 to 40.

IDN

**Sebaran Data**: Mayoritas data pasien tersebar di area di mana BMI berkisar antara 20 dan 40, kadar glukosa sekitar 100 hingga 150, dan usia berkisar antara 20 hingga 60 tahun.

**Korelasi**: Tidak ada korelasi linear atau langsung yang jelas antara BMI, usia, dan kadar glukosa di antara pasien diabetes ini. Seiring bertambahnya usia, tidak ada pola yang menunjukkan bahwa BMI atau kadar glukosa meningkat atau menurun secara signifikan.

**Konsentrasi Pasien**: Sebagian besar pasien memiliki kadar glukosa antara sekitar 100 hingga 150. Proporsi besar dari pasien ini berusia antara sekitar 20 hingga 60 tahun dengan BMI berkisar sekitar 20 hingga 40.
